# College GPA and SAT Scores

```python
# Run this cell first to load the numpy and matplotlib.pyplot modules
import numpy as np
# load graphics
import matplotlib.pyplot as plt
%matplotlib inline
from scipy import stats
```

```python
#Importing the different data
GPA = np.loadtxt('GPA.dat')
GPA_1 = np.loadtxt('GPA_1.dat')
GPA_2 = np.loadtxt('GPA_2.dat')
SAT = np.loadtxt('SAT.dat')
SAT_1 = np.loadtxt('SAT_1.dat')
SAT_2 = np.loadtxt('SAT_2.dat')
College = np.loadtxt('College.dat')
```

## Correlation

```python
#Overall data
plt.scatter(GPA,SAT, marker='.')
m,b = np.polyfit(GPA, SAT, 1)
gpa = np.linspace(0,5,100)
SAT_y = m*gpa+b
plt.plot(gpa,SAT_y)
plt.xlim(2,5)
r_SATGPA_m = np.corrcoef(GPA,SAT)
r_SATGPA = r_SATGPA_m[0,1]
plt.title('All Sampled Students')
plt.xlabel('GPA')
plt.ylabel('SAT Score')
```

```python
print('A linear regression was performed on the data set resulting in a regression line of the equation:')
print('SAT =',m,'* GPA +', b)
print('This line had a correlation coefficient of', r_SATGPA, 'and a coefficient of determination of ',r_SATGPA**2)
```

## Assessing Normality of SAT Scores

```python
SAT_avg = np.mean(SAT)
SAT_std = np.std(SAT)
plt.vlines(SAT_avg,0,200, color='red')
plt.vlines(SAT_avg+SAT_std,0,200, color='red', linestyle=':')
plt.vlines(SAT_avg-SAT_std,0,200, color='red', linestyle=':')
plt.hist(SAT, bins=30)
plt.ylim(0,150)
plt.xlim(600,1500)
plt.title('Histogram of all Sampled SAT Scores')
plt.xlabel('SAT Score')
plt.ylabel('Number of Students')
```

```python
#Running a Shapiro-Wilk test to determine if the data fits a normal distribution
print('The average SAT score of all students sampled is',SAT_avg, 'The standard deviatoin of the sample is',SAT_std)
SAT_SW_Test = stats.shapiro(SAT)
SAT_SW_p = SAT_SW_Test[1]
print('A Shapiro-Wilk test to asses the normality of the data was run, finding a corresponding p-value of',SAT_SW_p)
if SAT_SW_p < .005:
    print('The data does not appear to fit a normal distribution (alpha level = .005).')
else:
    print('The data appears to fit a normal distribution (alpha level = .005).')
```

```python
SAT_1_avg = np.mean(SAT_1)
SAT_1_std = np.std(SAT_1)
plt.vlines(SAT_1_avg,0,200, color='red')
plt.vlines(SAT_1_avg+SAT_1_std,0,200, color='red', linestyle=':')
plt.vlines(SAT_1_avg-SAT_1_std,0,200, color='red', linestyle=':')
plt.hist(SAT_1, bins=30)
plt.ylim(0,80)
plt.xlim(600,1500)
plt.title('Histogram of College 1 SAT Scores')
plt.xlabel('SAT Score')
plt.ylabel('Number of Students')
```

```python
#Running a Shapiro-Wilk test to determine if the data fits a normal distribution
print('The average SAT score of students attending college 1 is',SAT_1_avg, 'The standard deviatoin of the sample is',SAT_1_std)
SAT_1_SW_Test = stats.shapiro(SAT_1)
SAT_1_SW_p = SAT_1_SW_Test[1]
print('A Shapiro-Wilk test to asses the normality of the data was run, finding a corresponding p-value of',SAT_1_SW_p)
if SAT_1_SW_p < .005:
    print('The data does not appear to fit a normal distribution (alpha level = .005).')
else:
       print('The data appears to fit a normal distribution (alpha level = .005).')
```

```python
SAT_2_avg = np.mean(SAT_2)
SAT_2_std = np.std(SAT_2)
plt.vlines(SAT_2_avg,0,200, color='red')
plt.vlines(SAT_2_avg+SAT_2_std,0,200, color='red', linestyle=':')
plt.vlines(SAT_2_avg-SAT_2_std,0,200, color='red', linestyle=':')
plt.hist(SAT_2, bins=30)
plt.ylim(0,60)
plt.xlim(700,1200)
plt.title('Histogram of College 2 SAT Scores')
plt.xlabel('SAT Score')
plt.ylabel('Number of Students')
```

```python
#Running a Shapiro-Wilk test to determine if the data fits a normal distribution
print('The average SAT score of students attending colloege 2 is',SAT_2_avg, 'The standard deviatoin of the sample is',SAT_2_std)
SAT_2_SW_Test = stats.shapiro(SAT_2)
SAT_2_SW_p = SAT_2_SW_Test[1]
print('A Shapiro-Wilk test to asses the normality of the data was run, finding a corresponding p-value of',SAT_2_SW_p)
if SAT_2_SW_p < .005:
    print('The data does not appear to fit a normal distribution (alpha level = .005).')
else:
    print('The data appears to fit a normal distribution (alpha level = .005).')
```

## Two-Sample t-test to Assess Statistical Differences in SAT Scores

```python
SAT_T_t, SAT_T_p = stats.ttest_ind(SAT_1, SAT_2, equal_var=False)
#The p-value must be divided by 2, as the function for t-test is two tailed as opposed to one-tailed
print('The t-test resulted in a p-value of',SAT_T_p/2)
if SAT_T_p < .005:
    print('The SAT scores of students attending colloege 1 are significantly greater than those attending college 2 (alpha level = .005).')
else:
    print('The SAT scores of students attending college 1 are not significantly greater than those attending college 2 (alpha level = .005).')
```
